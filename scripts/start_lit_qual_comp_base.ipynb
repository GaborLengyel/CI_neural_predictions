```python
def convert_to_velocity(speed, angle_degrees):
    # Convert the angle from degrees to radians
    angle_radians = np.radians(angle_degrees)
    
    # Calculate velocity components
    v_x = speed * np.cos(angle_radians)
    v_y = speed * np.sin(angle_radians)
    
    return v_x, v_y

def velocity_to_speed_angle(v_x, v_y):
    # Compute the speed (magnitude of the vector)
    speed = np.sqrt(v_x**2 + v_y**2)
    
    # Compute the angle in radians and convert to degrees
    angle_radians = np.arctan2(v_y, v_x)
    angle_degrees = np.degrees(angle_radians)
    
    return speed, angle_degrees

def subtract_angles(angle1, angle2):
    result = (angle1 - angle2) % 360
    if result > 180:
        result -= 360
    return result
```
